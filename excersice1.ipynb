```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.random.rand(5, 1) * 10
y = 12.4356 * x - 6.28761
y = y + np.random.normal(scale=1, size=y.shape)
dataset5 = np.hstack((x, y))
```

```python
x = np.random.rand(25, 1) * 10
y = 12.4356 * x - 6.28761
y = y + np.random.normal(scale=1, size=y.shape)
dataset25 = np.hstack((x, y))
```

```python
x = np.random.rand(100, 1) * 10
y = 12.4356 * x - 6.28761
y = y + np.random.normal(scale=1, size=y.shape)
dataset100 = np.hstack((x, y))
```

```python
from sklearn.linear_model import LinearRegression
x = np.random.rand(1000, 1) * 10
y = 12.4356 * x - 6.28761
y = y + np.random.normal(scale=1, size=y.shape)
dataset1000 = np.hstack((x, y))
dataset=[dataset5,dataset25,dataset100,dataset1000]
model1 = LinearRegression(normalize=True)
coefs1=np.zeros((4,1))
scores1=np.zeros((4,1))
intercept1=np.zeros((4,1))
for i in range(0,4):
    x = dataset[i][:,0]
    y = dataset[i][:,1]
    x = x.reshape(-1, 1)
    y = y.reshape(-1, 1)
    model1.fit(x,y)
    x_line = np.arange(-3, 12, 0.1).reshape(-1, 1)
    y_line = model1.predict(x_line)
    coefs1[i]=model1.coef_
    intercept1[i]=model1.intercept_
    plt.subplot(4,1,i+1)
    plt.plot(x[:,0], y[:,0], 'b.')
    plt.plot(x_line, y_line, 'r--')
    scores1[i]=model1.score(x,y)
print(coefs1)
print(intercept1)
print(scores1)
```

```python
from sklearn.linear_model import LinearRegression
model2 = LinearRegression(normalize=True)
coefs2=np.zeros((4,4))
scores2=np.zeros((4,1))
intercept2=np.zeros((4,1))
for i in range(0,4):
    x = dataset[i][:,0]
    y = dataset[i][:,1]
    x = x.reshape(-1, 1)
    y = y.reshape(-1, 1)
    x = np.hstack((x, x ** 2, x ** 3, x ** 4))
    model2.fit(x,y)
    x_line = np.arange(-3, 12, 0.1).reshape(-1, 1)
    x_line = np.hstack((x_line, x_line ** 2, x_line ** 3, x_line ** 4))
    y_line = model2.predict(x_line)
    coefs2[i]=model2.coef_
    intercept2[i]=model2.intercept_
    plt.subplot(4,1,i+1)
    plt.plot(x[:,0], y[:,0], 'b.')
    plt.plot(x_line[:,0], y_line, 'r--')
    scores2[i]=model2.score(x,y)
print(coefs2)
print(intercept2)
print(scores2)
```
